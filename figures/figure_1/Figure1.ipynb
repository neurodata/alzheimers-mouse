```python
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from scipy.spatial.distance import pdist, squareform
```

```python
n = 5
lw = 3
```

```python
rng = np.random.default_rng(0)
x = rng.poisson(lam=3, size=(n, 1))

fig, ax = plt.subplots(figsize=(2, 2), dpi=300, layout='constrained')

sns.heatmap(x, square=True, linewidth=lw/2, linecolor='k', ax=ax, cbar=False, cmap='Blues')
sns.despine(ax=ax, left=True, bottom=True)


# Hide X and Y axes label marks
ax.xaxis.set_tick_params(labelbottom=False)
ax.yaxis.set_tick_params(labelleft=False)

# Hide X and Y axes tick marks
ax.set_xticks([])
ax.set_yticks([])

# Make border around heatmap
ax.axhline(y=0, color='k',linewidth=lw)
ax.axhline(y=n, color='k',linewidth=lw)
ax.axvline(x=0, color='k',linewidth=lw)
ax.axvline(x=.99, color='k',linewidth=lw)

ax.set_ylabel("# of ROIs")

fig.savefig('blue_col.pdf', bbox_inches='tight', pad_inches=0, transparent=True)
```

```python
dist = squareform(pdist(x, 'euclidean'))

fig, ax = plt.subplots(figsize=(2, 2), dpi=300, layout='constrained')

sns.heatmap(dist, square=True, linewidth=lw/2, linecolor='k', ax=ax, cbar=False, cmap='Reds')
sns.despine(ax=ax, left=True, bottom=True)


# Hide X and Y axes label marks
ax.xaxis.set_tick_params(labelbottom=False)
ax.yaxis.set_tick_params(labelleft=False)

# Hide X and Y axes tick marks
ax.set_xticks([])
ax.set_yticks([])

# Make border around heatmap
ax.axhline(y=0, color='k',linewidth=lw)
ax.axhline(y=n, color='k',linewidth=lw)
ax.axvline(x=0, color='k',linewidth=lw)
ax.axvline(x=n, color='k',linewidth=lw)

ax.set_ylabel("# of ROIs")
ax.set_xlabel("# of ROIs")

fig.savefig('dists1.pdf', bbox_inches='tight', pad_inches=0, transparent=True)
```
